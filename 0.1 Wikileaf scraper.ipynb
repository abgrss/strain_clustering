```python
import requests
import bs4 as BeautifulSoup
import numpy as np
import pandas as pd
# import re
```

```python
def page_scraper(url,df):
    '''
    Scrapes each Wikileaf page for each url. Returns a dataframe with the columns of df. 
    '''
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup.BeautifulSoup(plain_text, 'html.parser')

# GET STRAIN
    strain = soup.find("h1", {"itemprop" : "name"}).text
    print(strain)

# GET STRAIN_TYPE
    temp = strain.find_next('span')
    type_string = temp.text
    
#     str(soup.find("div", {"class" : "product-info-line cannabis"})) DELETE
    if "Indica" in type_string:
        strain_type = "indica"
    elif "Sativa" in type_string:
        strain_type = "sativa"
    elif "Hybrid" in type_string:
        strain_type = "hybrid"
    else:
        strain_type = "ERROR"

# GET STRAIN_TYPE PERCENTAGE
# We want to get the portion of type_string between </a> and </div>
    percentage = type_string[type_string.find("</a>")+4:type_string.rfind("</div>")] # redo as a regex?
    if " - " in percentage:
        percentage = percentage[3:] # removes " - "

    graph_values = soup.find_all(attrs={"class":"graph-val"})

    try:
        THC_max = str(graph_values[1])
        THC_max = THC_max[THC_max.find(">")+1:THC_max.rfind("</div>")]

# GET THC_avg (avgerage tested THC)
        THC_avg = str(graph_values[2])
        THC_avg = THC_avg[THC_avg.find(">")+1:THC_avg.rfind("</div>")]

    except:
        print(f'ERROR for {strain}, graph_values: {graph_values}')
        THC_max = 'ERROR'
        THC_avg = 'ERROR'
        

    data = pd.DataFrame([[strain, strain_type, percentage, THC_max, THC_avg]], columns = df.columns)

    return data
```

```python
def get_wikileaf_urls(master_list):
    '''
    Gets a list of each strain per page in the database
    '''
    
    page_list = []
    temp_num = 0
    max_val = 1500
    dummy_page = "https://www.wikileaf.com/strains/?result=" # Page 1 - ends at "1450"

# GET ALL PAGES IN THE WIKILEAF DB
    while int(temp_num) < max_val:
        page = dummy_page+str(temp_num)
        temp_num = temp_num + 50 # 50 strains shown per page
        print(page)
        page_list.append(page)
    
    print(page_list)

# GET STRAIN URLS FOR EACH PAGE IN THE LIST    
    for page in page_list:
        source_code = requests.get(page)
        plain_text = source_code.text
        soup = BeautifulSoup.BeautifulSoup(plain_text, 'html.parser')

        all_links = soup.find_all(attrs={"class":"item-title"})
    
        for html in all_links:
            html = str(html)
            link = html[html.find("href=")+6:html.rfind('\"><div class=\"')] # redo as a regex?

            master_list.append(link) # ADD THE STRAIN URL TO THE MASTER LIST
            print(master_list[-1])
    return master_list
```

```python
# RUN get_wikileaf_urls TO GET ALL THE URLS IN THE DB
master_list = []
get_wikileaf_urls(master_list)
```

```python
# Set up dataframes to scrape into
df = pd.DataFrame(columns=['Strain','Type','Pct','THC_max','THC_avg'])
```

```python
# FINALLY, FOR EACH URL IN master_list, SCRAP THE PAGE AND APPEND TO df
for url in master_list:
    data = page_scraper(url,df)
    df = df.append(data, ignore_index=True)
df
```
