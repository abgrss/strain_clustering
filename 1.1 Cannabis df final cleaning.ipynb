### Flavours

Despite looking like a list this is just a string. We can get each string, parse them, and add to a list or a set.

```python
flavours = set()

for i, item in enumerate(df['Flavours']):
    temp = item # Get flavour list (actually a string)
    temp = temp[1:-1]
    temp= temp.replace("'","")
    temp= temp.replace(",","")
    # print(temp)
    templist = temp.split() # Get every flavour from the row index
    # print(templist, type(templist))
    for flavour in templist:
        flavours.add(flavour)
print((flavours), len(flavours))
```

Great! 48 Flavours, of highly varying frequency (and relevance) no doubt.

Let's now use this set to manually one-hot encode flavours. First, let's put our set into an array.

```python
flav_array = np.array(list(flavours))
print(flav_array, len(flav_array))
flav_array.resize()
```

Now let's populate the corresponding values of flav_df.

```python
# Add the empty columns with flavour names
for flavour in flavours:
    df[flavour] = np.NaN  
df.shape
```

```python
# Populate each individual flavour column according to values in "Flavour"
for i, item in enumerate(df['Flavours']):
    print(i)
    temp = item # Get flavour list (actually a string)
    temp = temp[1:-1]
    temp= temp.replace("'","")
    temp= temp.replace(",","")
    # print(temp)
    templist = temp.split() # Get every flavour from the row index
    # print(templist, type(templist))
    for flavour in templist: # Make column value = 1 for each flavour in the cell
        df[flavour][i] = 1
```

```python
# Let's do a quick visualization of the total flavours inour df.
flavour_counts = {}
for column in df:
    if column in flavours:
        flavour_counts[column]=int(df[column].sum())

plt.figure(figsize=(12,8))
sns.barplot(y=list(flavour_counts.keys()), x=list(flavour_counts.values()))
# plt.barh(flavour_counts.values(), flavour_counts.keys())
plt.xlabel('Counts')
plt.title('Flavour counts in df')
plt.show();
```

### Effects, Medical, and Negatives

```python
df['Effmedneg'][0]
```

```python
# Get the set of "Effects", "Medical" (applications), and "Negatives"
effmedneg = set()

for i, item in enumerate(df['Effmedneg']):
    temp = item # Get Effmedneg list (actually a string)
    temp = temp[1:-1]
    temp = temp.replace("'","")
    temp = temp.replace(",","")
    templist = temp.split() # Split the effects and their values
#     print(templist)
    
    for thing in templist:
        if thing[0].isalpha():
            effmedneg.add(thing)
print(effmedneg, len(effmedneg)) 
```

Here our Eff/Med/Neg categories are all jumbled up and some of the compound terms are broken up, but we can fix all this later on.

Now, let's create corresponding effmedneg columns and populate them.

```python
# Add the empty columns with flavour names
for thing in effmedneg:
    df[thing] = np.NaN  
df.shape
```

```python
# Populate each individual effmedneg column according to their values
for i, item in enumerate(df['Effmedneg']):
    print(i)
    temp = item # Get Effmedneg list (actually a string)
    temp = temp[1:-1]
    temp = temp.replace("'","")
    temp = temp.replace(",","")
    templist = temp.split() # Split the effects and their values
    print(templist)
    
    eff_temp = "" # set blank eff_temp
    for item in templist: # Make column value = 1 for each flavour in the cell
        if item[0].isalpha(): # If the list item is a term (not a number)
            eff_temp = item
        else: # If the list item is a number (following an eff_temp)
            df[eff_temp][i] = float(item)/100 # divide by 100 to normalize between 0 and 1
#         print(i, item)
```

Let's do a quick column cleanup for compound names.
1. Delete "Muscle" - change "Spasms" to "Muscle_spasms"
2. Delete "Dry" - change "Mouth" to "Dry_mouth", "Eyes" to "Dry_eyes"
3. Delete "Lack", "of" - change "Appetite" to "Lack_of_appetite"
3. Delete "Eye" - change "Pressure" to "Eye_pressure"

```python
df.rename(columns={'Spasms': 'Muscle_spasms','Mouth' : "Dry_mouth", "Eyes" : "Dry_eyes",\
                  "Appetite" : "Lack_of_appetite", "Pressure": "Eye_pressure"}, inplace=True)
df.drop(columns=['Muscle', 'Dry','Lack','Eye','of'], inplace=True)
df.columns
```

```python
# Don't forget to update and remove these items from the set effmedneg 
effmedneg.difference_update({'Muscle', 'Dry','Lack','Eye','of'})
effmedneg.difference_update({'Spasms','Mouth','Eyes','Appetite','Pressure'})
effmedneg.update({'Muscle_spasms', 'Dry_mouth', 'Dry_eyes', 'Lack_of_appetite', 'Eye_pressure'})
print(effmedneg, len(effmedneg))
```

```python
# Let's do a quick visualization of the total effmednegs in our df.
effmedneg_counts = {}
for column in df:
    if column in effmedneg:
        effmedneg_counts[column]=int(df[column].sum())

plt.figure(figsize=(12,8))
sns.barplot(y=list(effmedneg_counts.keys()), x=list(effmedneg_counts.values()))
# plt.barh(flavour_counts.values(), flavour_counts.keys())
plt.xlabel('Counts')
plt.title('Effects/Medical/Negatives counts in df')
plt.show();
```

Excellent! Let's now categorize these and add prefixes: effects (Eff_) and Medical (Med_) - Effects are the outcomes of ingesting cannabis, while Med_ are the symptoms people are attempting to treat by taking cannabis.

```python
# Make a set of just those medical conditions
med_set = set({'Cramps', 'Depression', 'Eye_pressure', 'Fatigue', 'Headaches',\
              'Inflammation', 'Insomnia' , 'Lack_of_appetite', 'Muscle_spasms',\
              'Nausea', 'Pain', 'Seizures', 'Spasticity', 'Stress'})
```

```python
for column in df:
    if column in effmedneg: # If an effmedneg column        
        if column in med_set: # If a medical condition
            new_name = "Med_" + column
            df.rename(columns={column : new_name}, inplace=True)
        else:
            new_name = "Eff_" + column
            df.rename(columns={column : new_name}, inplace=True)
df.columns
```

Now we can visualize our effects and medical symptoms separately.

```python
# Visualze effects
effects = {}
medicals = {}
for column in df:
    if "Eff_" in column:
        effects[column]=int(df[column].sum())
    if "Med_" in column:
        medicals[column]=int(df[column].sum())

plt.figure(figsize=(12,8))
sns.barplot(y=list(effects.keys()), x=list(effects.values()))
plt.xlabel('Counts')
plt.title('Effect counts in our df')
plt.show();

plt.figure(figsize=(12,8))
sns.barplot(y=list(medicals.keys()), x=list(medicals.values()))
plt.xlabel('Counts')
plt.title('Medical symptom counts in our df')
plt.show();
```
