```python
import requests
import bs4 as BeautifulSoup
import numpy as np
import pandas as pd
# import re
```

```python
# Set up dataframes to scrape into
df = pd.DataFrame(columns=['Strain','Type_hyt','Description_hyt','THC_hyt'])
```

```python
def page_scraper(url,df):
    '''
    Scrapes each Hytiva.com page for each url. Returns a dataframe with the columns of df. 
    '''
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup.BeautifulSoup(plain_text, 'html.parser')

# GET STRAIN & STRAIN_TYPE
    strain = ""
    strain_type = ""
    try:
        temp = soup.find("title", {"itemprop" : "name"}).text
        strain = temp[:temp.find(" - ")]
        strain_type = temp[temp.find(" - ")+3:temp.find(" Cannabis")].lower()
    except:
        print(f"COULD NOT GET STRAIN OR STRAIN_TYPE NAME FOR {strain}")
        strain = "ERROR"
    print(f"{strain}\n{strain_type}")
    

# GET DESCRIPTION
    description = ""
    try:
        temp = ""
        description_bowl = soup.find("div", {"class" : "Detail Body"})
#         print(description_bowl)
        plist = description_bowl.find_all("p")
        for item in plist:
            description = description + item.text + " "
    except:
        print(f"COULD NOT GET DESCRIPTION FOR {strain}")
        description = "ERROR"
#     print(description)

# GET THC
    THC = ""
    try:
        tempbowl = []
        tempbowl = soup.find_all("div", {"class" : "Column One"}) # First get "Column One"
        for item in tempbowl:
            if "Cannabinoid Lab Analysis" in item.text: # find the soup element with a THC value (if it exists)
                tds = item.find_all("td") # scrape every element of the Lab Analysis into a list
                tds_text = [td.text for td in tds] # convert each element in the list to a string
                THC_cell = tds_text.index("THC:") # Find the cell containing THC
                THC = tds_text[THC_cell+1]
                break
    except:
        print(f"COULD NOT GET THC FOR {strain}")
        THC = "ERROR"   
        
    data = pd.DataFrame([[strain, strain_type, description, THC]], columns = df.columns) # 'Description_hyt','THC_hyt'
    print(data)
    return data
```

```python
def get_hytiva_urls(master_list):
    '''
    Gets a list of each strain per page in the database
    '''
    
    page_list = []
    temp_num = 1
    max_val = 58
    dummy_page = "https://www.hytiva.com/strains?page=" # Page 1 - ends at "1450"
    
    
    
# GET ALL PAGES IN THE HYTIVA DB
    while int(temp_num) < max_val:
        page = dummy_page+str(temp_num)
#         temp_num = temp_num + 24 # 50 strains shown per page
        print(page)
        page_list.append(page) # Call pagelist function - appends to master_list)
        temp_num = temp_num + 1

        
    print(page_list)

# GET STRAIN URLS FOR EACH PAGE IN THE LIST    
    for page in page_list:
        source_code = requests.get(page)
        plain_text = source_code.text
        soup = BeautifulSoup.BeautifulSoup(plain_text, 'html.parser')

        all_bowls = soup.find_all(attrs={"class":"Listing-Title Listing-Link"})
    
        for link in all_bowls:
            link = link.get("href")
            link = str(link)
#             link = html[html.find("href=")+6:html.rfind('\"><div class=\"')]
            master_list.append(link) # ADD THE STRAIN URL TO THE MASTER LIST
            print(master_list[-1])
    return master_list
```

Now, we create our master_list and call get_hytiva_urls. This scrape all strain pages and return a list of all strain url substrings.

```python
master_list = []
get_hytiva_urls(master_list)
```

Now, for each url substring in master_list, scrape the page and append it to a new df.

```python
df = pd.DataFrame(columns=['Strain','Type_hyt','Description_hyt','THC_hyt'])

# transform the url substrings in master_list to full urls, appends to url_list
url_list = []
for url in master_list:
    url_list.append("https://www.hytiva.com/" + url)
# [print(url) for url in url_list]

counter = 1
for url in url_list:
    data = page_scraper(url,df)
    df = df.append(data, ignore_index=True)

    print(counter)
    counter = counter + 1
df
```
