### Create an indica/sativa scale

```python
# Now for every Pct that's not a NaN, let's get the Strain_type and merge then
# Before merging, flip all values for sativa

lean_wiki = []

for row in range(0,len(df)): # Since all the values in Pct are strings except for NaNs, we can select for strings.
    print()
    # lean is the sativa-indica score - lean=0 if 100% sativa, 1 if 100% indica, and 0.5% if an evently balanced hybrid
    lean = np.NaN # Since we're creating a series we will append a NaN if no values present
    if type(df.Strain_type[row]) == str:
        if type(df.Pct[row]) == str: # if both Strain_type and Pct are strings (not NaN)

            if "hybrid" in df.Strain_type[row]:
                lean = 0.5 # "hyrbid" is always 50% in the wikileaf df
                
            elif "indica" in df.Strain_type[row]:
                if "100" in df.Pct[row]: # 100% indica -> 1
                    lean = 1.0
                else:
                    lean = float(df.Pct[row][0:2])/100
                    
            elif "sativa" in df.Strain_type[row]:
                if "100" in df.Pct[row]: # 100% sativa -> 0
                    lean = 0.0
                else:
                    temp = float(df.Pct[row][0:2])/100
                    lean = 1-temp
            else:
                print(f"ERROR. Strain_type: {df.Strain_type[row]} PCT: {df.Pct[row]}")
    lean_wiki.append(lean)
```

```python
# Now let's find the mean values for sativa-dominant (1-49) and indica-dominant (51-99) plants

import math
sativa_values_ci = []
indica_values_ci = []

for i in range(len(lean_ci)):
    temp = lean_ci[i]
    if type(temp) == float:
        if (temp > 0.0) & (temp < 0.5):
            sativa_values_ci.append(temp)
        elif (temp > 0.5) & (temp < 1.0):
            indica_values_ci.append(temp)

# set means
sativa_mean_ci = np.average(sativa_values_ci)
indica_mean_ci = np.average(indica_values_ci)

# set standard deviations
sativa_std_ci = np.std(sativa_values_ci)
indica_std_ci = np.std(indica_values_ci)

print(f"sativa mean: {sativa_mean_ci}, indica mean: {indica_mean_ci}, \nsativa std: {sativa_std_ci}, indica std: {indica_std_ci}")
```

Now let's create two small functions to return a truncated normal distribution for our sativa and indica values.

```python
from scipy.stats import truncnorm

def truncated_normal(mean=0, sd=1, min=0, max=10):
    return truncnorm((min - mean) / sd, (max - mean) / sd, loc=mean, scale=sd)

sativa_trunc_func = truncated_normal(mean=sativa_mean_ci, sd=sativa_std_ci, min=0, max=0.5)
indica_trunc_func = truncated_normal(mean=indica_mean_ci, sd=indica_std_ci, min=0.5, max=1)
```

```python
# Let's check sativa_trunc_func and indica_trunc_func
sat_temp = []
ind_temp = []
for i in range(0,10000):
    sat_temp.append(sativa_trunc_func.rvs())
    ind_temp.append(indica_trunc_func.rvs())
    
plt.hist(sat_temp)
plt.hist(ind_temp)
print(np.average(sat_temp), min(sat_temp), max(sat_temp))
print(np.average(ind_temp), min(ind_temp), max(ind_temp))
```

All meaningful data here is a string. There are two types of data in one here: caregorical and numerical. Since we have so much numerical data (especially from wikileaf), let's just use that.

```python
ci_values = []
for item in df.THC_ci:
    if type(item) == float: # If item is a NaN
        ci_values.append(item)
    else:
        if "%" in item or "-" in item: # All numbers
            temp = str(item)
            temp = temp.replace("%","") # Cleaning
            temp = temp.replace("+","")
            if "-" in temp: # if a range - get the average
                num1 = float(temp[0:temp.find("-")])
                num2 = float(temp[temp.find("-")+1:])
                THC_value = (num1+num2)/2 # The average of the range
            else: # if value is a single number (not a range)
                THC_value = float(temp)
                print(THC_value)
        ci_values.append(THC_value)

values_nonans = [i for i in ci_values if str(i) != 'nan']
meta_average = np.mean(values_nonans)
print(f"Average of 'values': {meta_average}")
print(f"Length of ci_values: {len(ci_values)}")

plt.figure()
# plt.hist(averages,20)
sns.distplot(ci_values,20)
plt.xlabel('Numerical THC percentages')
plt.ylabel('proportion of total')
plt.title('THC values for cannabis.info')
plt.show();
```

After having inspected the wikileaf and cannabis-info databases more closely, I have decided that for any value difference over 10 (ABS) I will go with wikileaf's value, and that for any difference under 10 (ABS) I'll take the average of the two.

LOGIC:
- check ci -> if nan, check wiki -> if nan, then THC == NAN -> else, THC == wiki
- 
- check wiki -> if nan, check ci.
- if wiki is not nan


```python
# Now, create the THC column
THC = []
for i, ci_value in enumerate(ci_values):
    if math.isnan(ci_values[i]): # If ci is NaN
        if math.isnan(wiki_values[i]): # If wiki is NaN
            THC.append(np.NaN) # append NaN (both NaN)
        else: # If wiki is a number
            THC.append(wiki_values[i])
    
    else: # If ci is a number
        if math.isnan(wiki_values[i]): # If wiki is NaN
            THC.append(ci_values[i]) # append ci
        else: # Both numbers:
            if abs(wiki_values[i] - ci_values[i]) > 10: # If value difference is larger than 10...
                THC.append(wiki_values[i]) # ... append wikileaf's value
                print(wiki_values[i])
            else: # If value difference is under 10...
                THC.append((wiki_values[i] + ci_values[i])/2) # ... append the average of the two values

df['THC'] = pd.DataFrame(data=THC)

# Show the df for no NAs
# df[df['THC'].notna()].sort_values(by="THC", ascending=False)
print(f"Total # of values for THC: {len(df[df['THC'].notna()])}")
```

```python
# Finally, we clean the rows, making them numerical lean values.
# We'll have to flip the values if an "s" is found after the ratio.

for i, row in enumerate(df['Lean_leafly']):
    if type(row) == str: # If not a NaN
        ratio = row[0:5]
        strain_type = row[-1] # i(ndica), s(ativa)

        if strain_type == "s": # Because of the way lean is, flip the values if strain is "s"!
            ratio = ratio[3:5]+"/"+ratio[0:2]
#         print(ratio)
        lean = (float(ratio[0:2])/100)
        df['Lean_leafly'][i] = lean
```
