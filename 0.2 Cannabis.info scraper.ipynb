```python
import requests
import bs4 as BeautifulSoup
import numpy as np
import pandas as pd
# import re
```

```python
def get_cannabis_info_urls(master_list):
    '''
    Gets a list of each strain per letter/0-9 in the database
    '''

# GET ALL PAGES IN THE cannabis.info DB
    page_list = []
    link = 'https://www.cannabis.info/en/strains?show_char='
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    for letter in alphabet:
        page_list.append(link+letter)
    page_list.append('https://www.cannabis.info/en/strains?show_char=0-9')    

# GET STRAIN URLS FOR EACH PAGE IN THE LIST
    for page in page_list:
        source_code = requests.get(page)
        plain_text = source_code.text
        soup = BeautifulSoup.BeautifulSoup(plain_text, 'html.parser')

        strain_links = soup.find("ul", {"class" : "strains-list"})

        for li in strain_links.findAll('li'):
            li = str(li)
            link = li[li.find("href=")+6:li.rfind('\">')] # redo as a regex?    
            print(link)
            master_list.append(link) # ADD THE STRAIN URL TO THE MASTER LIST

    return master_list
```

```python
def page_scraper(url,df):
    '''
    Scrapes page (url) to get the strain name, genetics, and THC level
    '''
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup.BeautifulSoup(plain_text, 'html.parser')

# GET STRAIN
    strain = soup.find("h1", {"itemprop" : "name"}).text
    print(strain)

# Find the first bit of relevant data: "Genetics"
    temp = soup.find("div", {"class" : "feature-title"})
#     print(temp)
    if temp.text == 'Genetics':
        genetics = temp.find_next("div")
        genetics = genetics.text.strip()
        print(genetics)
    else:
        print(f'ERROR: could not get genetics for strain {strain}')
        genetics = 'ERROR'

    try:
# Advance through the list to the next relevant data
        temp = temp.find_next("div", {"class" : "feature-title"}) # "Parents"
        temp = temp.find_next("div", {"class" : "feature-title"}) # "THC"
#         print('temp2:', temp)

        if temp.text == 'THC':
            THC = temp.find_next("div")
            THC = THC.text.strip()
            print(THC)
        else:
            print(f'ERROR: could not get THC for strain {strain}')
            THC = 'ERROR'
    except:
        THC = 'ERROR'
        
    data = pd.DataFrame([[strain, genetics, THC]], columns = df.columns)

    return data
```

```python
master_list  = []
get_cannabis_info_urls(master_list)
master_list
```

```python
# Set up dataframes to scrape into
df = pd.DataFrame(columns=['Strain','Genetics','THC'])
```

```python
##### FINALLY, FOR EACH URL IN master_list, SCRAP THE PAGE AND APPEND TO df
for url in strain_list:
    data = page_scraper(url,df)
    df = df.append(data, ignore_index=True)
df
```
